```python
from glob import glob
```

```python
img_list = glob("/content/drive/MyDrive/img/*.jpg") # * : all
```

```python
img_list
```

```python
import random
```

```python
random.shuffle(img_list)
img_list
```

```python
m = len(img_list)
m
```

```python
train_index = int(m*0.6)
val_index = int(m*0.2)
test_index = int(m*0.2)
```

```python
train_img_list = img_list[0:train_index]
val_img_list = img_list[train_index : train_index + val_index]
test_img_list = img_list[val_index : val_index + test_index]
```

```python
# 빈파일 생성
with open('/content/drive/MyDrive/digit_train.txt', 'w') as f:
  f.write('\n'.join(train_img_list) + "\n")

with open('/content/drive/MyDrive/digit_val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + "\n")

with open('/content/drive/MyDrive/digit_test.txt', 'w') as f:
  f.write('\n'.join(test_img_list) + "\n")
```
